<h3>Train-Test Splitting</h3>

```python
X = df_selected_cols1.drop("Driver At Fault", axis=1).values
y = df_selected_cols1["Driver At Fault"].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 21, stratify=y)
```

```python
print("X_train : ",X_train.shape)
print("X_test : ",X_test.shape)
print("y_train : ",y_train.shape)
print("y_test : ",y_test.shape)
```

<h3>KNN Classifier</h3>

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
knn_train_acc = []
knn_train_prec = []
knn_train_rec = []
knn_train_f1 =[]

knn_test_acc = []
knn_test_prec = []
knn_test_rec = []
knn_test_f1 =[]
```

</h4> It will some time to execute for each neigbhor value from [5,10,15,20,25]

```python
for i in range(5, 30, 5):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)


    knn_train_acc.append(knn.score(X_train,y_train))
    knn_test_acc.append(knn.score(X_test,y_test))

    y_train_pred = knn.predict(X_train)
    y_test_pred = knn.predict(X_test)

    ##print(confusion_matrix(y_test, y_test_pred))
    ##print(confusion_matrix(y_train, y_train_pred))

    ##print(classification_report(y_train, y_train_pred))
    ##print(classification_report(y_test, y_test_pred))

    train_report = classification_report(y_train, y_train_pred, output_dict=True)
    test_report = classification_report(y_test, y_test_pred, output_dict=True)

    knn_train_prec.append(train_report["1.0"]["precision"])
    knn_test_prec.append(test_report["1.0"]["precision"])

    knn_train_rec.append(train_report["1.0"]["recall"])
    knn_test_rec.append(test_report["1.0"]["recall"])

    knn_train_f1.append(train_report["1.0"]["f1-score"])
    knn_test_f1.append(test_report["1.0"]["f1-score"])

print("knn Train accuracies : ", knn_train_acc)  
print("knn Test accuracies : ", knn_test_acc) 

print("knn Train precisions : ", knn_train_prec)  
print("knn Test precisions : ", knn_test_prec)
```

<h3> Combining prediction from all models to check if we get more precision (kind of ensemble learning)</h3>

```python
#we have prediction from all the models as follows
y_test_pred #prediction from knn model
y_pred_log #predictions from logistic regression
y_nBtest_pred #predictions from naive bayesian
y_rftest_pred #predictions from random forest

#converting them to integers
y_test_pred = y_test_pred.astype('int32')
y_pred_log = y_pred_log.astype('int32')
y_nBtest_pred = y_nBtest_pred.astype('int32')
y_rftest_pred = y_rftest_pred.astype('int32')

#we will built new list with the following logic => new list will contain 1 only when corresponding all four prediction say 1
knn_y = list(y_test_pred)
log_y = list(y_pred_log)
nb_y = list(y_nBtest_pred)
rf_y = list(y_rftest_pred)

new_pred = []
```

```python
for i in range(0, len(knn_y)):
    if(knn_y[i] == 1 and log_y[i] == 1 and nb_y[i] == 1 and rf_y[i] == 1):
        new_pred.append(1)
    else:
        new_pred.append(0)
```
